```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Cargamos la Base de datos completa sin limpiar

```python
datos='/content/sample_data/AMZN.csv'
database = pd.read_csv(datos)

database
```

Aplicamos la funcion que creamos a la base de datos

```python
list(database)
```

```python
XDC = database[['Open', 'High', 'Low', 'Close', 'Volume']]
XDC = XDC.dropna()

XC = XDC[['Open', 'Low', 'Close', 'Volume']]

YC = XDC[['High']]

XC = np.array(XC)
YC = np.array(YC)

#NORMALIZO LOS DATOS
XCn = XC/np.max(XC, axis=0)
YCn = YC/np.max(YC, axis=0)
```

Ahora que ya esta la tabla lista para trabajar, comenzamos con la implementacion de los modelos

Modelo Madaline usando 2 capas, una con 10 nodos y otra con 1 nodo, ambas lineales

```python
NE = 4

modelM2=Sequential()
modelM2.add(Dense(10, activation='linear', use_bias=False, input_dim = NE))
modelM2.add(Dense(1, activation='linear'))

modelM2.compile(optimizer='adam', loss='mse')


modelM2.fit(XCn, YCn, epochs=200, batch_size=668)

Ypm2=modelM2.predict(XCn)
```

```python
#matriz de correlaciones
dfM2=pd.DataFrame(np.column_stack((YCn, Ypm2)))
dfM2.corr()
```

```python
#efectos individuales
W1M2 = modelM2.layers[0].get_weights()[0]
W2M2 = modelM2.layers[1].get_weights()[0]

np.dot(W1M2,W2M2)
```

implementamos el otro modelo, en este caso un modelo logit sencillo de solo 1 capa

```python
NE = 4

modelL=Sequential()
modelL.add(Dense(1, activation='sigmoid', use_bias=False, input_dim = NE))

modelL.compile(optimizer='adam', loss='mse')


modelL.fit(XCn, YCn, epochs=200, batch_size=668)

YpL=modelL.predict(XCn)
```

```python
#matriz de correlaciones
dfa=pd.DataFrame(np.column_stack((YCn, YpL)))
dfa.corr()
```

```python
#efectos individuales
WL = modelL.layers[0].get_weights()
WL
```

Ahora hacemos los pronosticos

```python
#pronostico del high de la accion segun cada modelo
Ypm2
YpL
#creamos el factor, se toma el high original, en este caso seria el YC
factor = np.min(YC)+(np.max(YC)-np.min(YC))

#pronostico score madaline
Ypm2_desnormalizado = Ypm2*factor

#adaline
YpL_desnormalizado = YpL*factor
```

```python
#pronosticamos el valor mas alto de una accion ingresando el open, min, cierre y el volumen que se vendio de la accion que queremos evaluar
Accion_evaluar = np.array([[0.098438, 0.085417, 0.086458, 294000000]])
Accion_evaluar_normalizada = Accion_evaluar/np.max(XC, axis = 0)

#pronostico individuo con modelo lineal 2 capas
Prediccion_maximo_accion_M2 = modelM2.predict(Accion_evaluar_normalizada)
Prediccion_maximo_accion_desnormalizado_M2 = Prediccion_maximo_accion_M2*factor

print('El valor maximo de la accion que ingresaste es de:', Prediccion_maximo_accion_desnormalizado_M2)

#pronostico individuo con modelo combinado logit y lineal
Prediccion_maximo_accion_L = modelL.predict(Accion_evaluar_normalizada)
Prediccion_maximo_accion_desnormalizado_L = Prediccion_maximo_accion_L*factor

print('El valor maximo de la accion que ingresaste es de:', Prediccion_maximo_accion_desnormalizado_L)
```
